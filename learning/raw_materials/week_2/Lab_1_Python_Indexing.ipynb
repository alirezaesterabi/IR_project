# Part 2: Indexing

#### **Building a Vocabulary**

Lets start with 5 documents, where each document is just a simple sentence

```python
# documents to be indexed
documents = ['Boats are great',
             'Sailing boats are enviromental',
             'canal boats are narrow',
             'wooden boats are more work than metal boats',
             'a wooden boat is always the prettiest']

# documents combined
document_terms = [doc.split(' ') for doc in documents]
```

For demonstration purposes, we'll now build the index step by step. Note that there is a much quicker method which we'll show later.

```python
# First we need to build the "vocabulary" i.e all the words in the docs.

vocabulary = set()

for doc in documents:
  for word in doc.split():
    vocabulary.add(word)

vocabulary = list(vocabulary)

vocabulary
```

```python
# CODING TIP: You can also write this in a single line
vocabulary2 = list(set(word for doc in documents for word in doc.split()))

# Lets check that the two methods lead to the same result
vocabulary == vocabulary2
```

**Question**: Why do we build the vocabulary using a python set and not a list?

If you are struggling to understand why.. try to build the the vocabulary using a list instead of a set. What are the differences?

#### **Building a binary term-document matrix**

We can represent each document as a binary term-document matrix to show if a term features in that document or not

```python
# binary term-document matrix representation of all docs
binary_term_doc = [[ 1 if term in doc else 0 for term in vocabulary] for doc in document_terms]

binary_term_doc
```

```python
# CODING TIP: Alternatively you can also cast the boolean result of evaluating term in doc to an int
binary_term_doc2 = [[int(term in doc) for term in vocabulary] for doc in document_terms]

# Sanity check: Lets check that the two methods lead to the same result
binary_term_doc2 == binary_term_doc
```

```python
# Show in pandas for good visualisation
binary_term_doc_df = pd.DataFrame(binary_term_doc2, columns = vocabulary)
binary_term_doc_df
```

#### **Term Occurence and Document Frequency**

```python
# An important part of the index is a dictionary that tells us which documents contain which words, otherwise known as the term occurence dictionary. So let's create one
term_occurence = {term:[] for term in vocabulary}

for doc_id, doc in enumerate(document_terms):
  for term in doc:
    term_occurence[term].append(doc_id)
```

```python
term_occurence
```

```python
# CODING TIP: We can also use the defaultdict module from the collections package
# This dictionary implementation has a little more functionality, that may come in useful later.
from collections import defaultdict

term_occurence = defaultdict(list)

for doc_id, doc in enumerate(document_terms):
  for term in doc:
    term_occurence[term].append(doc_id)
```

```python
term_occurence
```

The **term occurence dictionary** can be used to calculate the **document frequency** for each term. i.e how many documents does each term appear in. This is an important measure that we'll see again in later labs

```python
doc_frequency = {keys: len(values) for (keys, values) in term_occurence.items()}
doc_frequency
```

#### **Term-document count matrix**

We have just explored code that calculates the binary term-document matrix. This representation is simple to calculate but it has its limitations.

What if some documents contain many occurences of the same term? It could be important to give more weight to these term when trying to retrieve the correct document. The binary term-document matrix cannot represent this information..

Lets introduce the **term-document count matrix** where we count the number of occurences of each vocab term in each document.

What differences do you expect to see?

```python
documents
```

```python
# binary term-document matrix we calculated early

binary_term_doc_df
```

```python
# term-document count matrix representation for all docs

# first create an matrix full of zeros
term_doc_count_matrix = np.zeros((len(document_terms), len(vocabulary)), dtype=int)

# iterate over the documents
for i, doc in enumerate(document_terms):
  # iterate over the terms in each document
  for term in doc:
    # find if a term is in our vocab
    if term in vocabulary:
      j = vocabulary.index(term) # find index of term in vocab
      term_doc_count_matrix[i, j] += 1

term_doc_count_df = pd.DataFrame(term_doc_count_matrix, columns = vocabulary)
term_doc_count_df
```

Spot the difference?

We mentioned previously that there was a quicker method for computing our term-document indices. It uses sklearn's **CountVectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# See https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
# binary = True returns the binary term-document matrix, while binary=False returns the term-document count matrix
vectorizer = CountVectorizer(binary=False)
```

```python
binary_term_doc = vectorizer.fit_transform(documents)
# a sparse matrix representaion of the data (lower memory consumption, especially useful for large documents). this is easy to turn into a numpy array with .toarray()
# however, don't do it unless you have to as it is costly
binary_term_doc
```

```python
vocabulary = vectorizer.get_feature_names_out()
vocabulary
```

```python
# visualise using pandas
df = pd.DataFrame(binary_term_doc.toarray(), columns=vocabulary)
df

# CODING TIP: numpy and pandas functions are often faster than using python's inbuilt iterators (for, while) due to vectorizing.
```
